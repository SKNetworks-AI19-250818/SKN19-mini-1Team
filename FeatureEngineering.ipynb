```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("./data/training/final/travel_insight.csv")
df
```

```python
# TRAVEL_STATUS_ACCOMPANY → Label Encoding
# _CODE 를 붙여서 추가 컬럼 생성
tsa = LabelEncoder()
df["TRAVEL_STATUS_ACCOMPANY_CODE"] = tsa.fit_transform(df["TRAVEL_STATUS_ACCOMPANY"])

# df["TRAVEL_STATUS_ACCOMPANY_CODE"]
print(df[["TRAVEL_STATUS_ACCOMPANY", "TRAVEL_STATUS_ACCOMPANY_CODE"]].head())
```

```python
# TRAVEL_MISSIONCHECK → One-Hot Encoding
# 너는(?) TRAVEL_MISSION_CHECK_(숫자) 로 컬럼 생성
tmc = pd.get_dummies(df["TRAVEL_MISSIONCHECK"], prefix="TRAVEL_MISSION_CHECK")
df = pd.concat([df, tmc], axis=1)

print(df.filter(like="TRAVEL_MISSION_CHECK").head())
```

```python
# MVMN_NM → Label Encoding
# _CODE 를 붙여서 추가 컬럼 생성
mnm = LabelEncoder()
df["MVMN_NM_CODE"] = mnm.fit_transform(df["MVMN_NM"])

print(df[["MVMN_NM", "MVMN_NM_CODE"]].head())
```

```python
# TRAVEL_START_YMD, TRAVEL_END_YMD → 연 / 월 분리
# TRAVEL_START_YEAR, TRAVEL_START_MONTH 로 컬럼 생성
# to_datetime으로 분할 가능
df["TRAVEL_START_YMD"] = pd.to_datetime(df["TRAVEL_START_YMD"])
df["TRAVEL_END_YMD"] = pd.to_datetime(df["TRAVEL_END_YMD"])

df["TRAVEL_START_YMD"] = pd.to_datetime(df["TRAVEL_START_YMD"])
df["TRAVEL_END_YMD"] = pd.to_datetime(df["TRAVEL_END_YMD"])

df["TRAVEL_START_YEAR"] = df["TRAVEL_START_YMD"].dt.year
df["TRAVEL_START_MONTH"] = df["TRAVEL_START_YMD"].dt.month
df["TRAVEL_END_YEAR"] = df["TRAVEL_END_YMD"].dt.year
df["TRAVEL_END_MONTH"] = df["TRAVEL_END_YMD"].dt.month

print(df[["TRAVEL_START_YMD", "TRAVEL_START_YEAR", "TRAVEL_START_MONTH"]].head())
print(df[["TRAVEL_END_YMD", "TRAVEL_END_YEAR", "TRAVEL_END_MONTH"]].head())
```
