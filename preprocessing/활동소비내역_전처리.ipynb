### # 4. 데이터 시각화를 통한 탐색

```python
# '활동유형(대분류) 비중 - ACTIVITY_TYPE_CD' 원형 그래프화

# 한글 폰트 (윈도우)
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False      # 음수 기호 깨짐 방지

df = tn_activity_consume_his_df
col = 'ACTIVITY_TYPE_CD'

# 라벨 매핑(있으면 채워서 사용, 없으면 숫자 그대로 표시)
type_map = {1:'취식', 2:'쇼핑/구매', 3:'체험 활동/입장 및 관람', 6:'기타'}  # 예시

# 빈도 집계 (NaN 포함하려면 .fillna('미상') 후 value_counts(dropna=False))
counts = df[col].value_counts()              # 많이 나온 순
# counts = df[col].value_counts().sort_index()  # 코드값 순으로 보고 싶으면

labels = [type_map.get(k, str(k)) for k in counts.index]
sizes  = counts.values
total  = sizes.sum()

# 작은 조각 강조(예: 6% 미만 조각만 explode)
min_pct_for_label = 6
explode = np.where(sizes / total * 100 < min_pct_for_label, 0.06, 0)

def autopct_fmt(p):
    # 작은 비율은 웨지 안쪽 표기 생략 → 범례로만 깔끔히 표시
    return f'{p:.1f}%' if p >= min_pct_for_label else ''

fig, ax = plt.subplots(figsize=(7,5))
wedges, _, autotexts = ax.pie(
    sizes,
    startangle=90,
    counterclock=False,
    labels=None,                           # 라벨은 범례에서 처리
    autopct=autopct_fmt,
    pctdistance=0.68,
    labeldistance=1.08,
    explode=explode,
    wedgeprops={'linewidth':1, 'edgecolor':'white'}  # 경계선으로 가독성↑
)
ax.axis('equal')
ax.set_title('활동유형(대분류) 비중 - ACTIVITY_TYPE_CD')

# 범례: 라벨 + 건수 + 비율
legend_labels = [f'{lab}  {cnt:,}건  ({cnt/total:.1%})'
                 for lab, cnt in zip(labels, sizes)]
ax.legend(wedges, legend_labels, title='범례',
          loc='center left', bbox_to_anchor=(1, 0.5))

# (옵션) 도넛 스타일 & 총 표본수 표기 — 원하면 주석 해제
# centre = plt.Circle((0,0), 0.55, fc='white')
# ax.add_artist(centre)
# ax.text(0, 0, f'N={total:,}', ha='center', va='center')

plt.tight_layout()
plt.show()
```

### 5. 데이터 정제 및 전처리

```python
# 드롭할 컬럼 목록만 정리
drop_cols = [
    'SGG_CD','ROAD_NM_ADDR','LOTNO_ADDR','ROAD_NM_CD','LOTNO_CD',
    'BRNO','PAYMENT_DT','CONSUME_HIS_SEQ','CONSUME_HIS_SNO'
]

# 원본은 보존하고, 없는 컬럼이 있어도 에러 없이 진행
df = tn_activity_consume_his_df.drop(
    columns=drop_cols, errors='ignore'
).copy()
```

```python
cols = ['STORE_NM', 'PAYMENT_ETC']

# 결측치 개수 확인(선택)
print("before:", {c: int(df[c].isna().sum()) for c in cols})

# NaN → "기타" 로 일괄 대체
df[cols] = df[cols].fillna('기타')
# (동일 의미) df.fillna({'STORE_NM':'기타', 'PAYMENT_ETC':'기타'}, inplace=True)

print("after:", {c: int(df[c].isna().sum()) for c in cols})
```

##### 데이터 정제 및 전처리 내역
- 제거
    - SGG_CD : 결측치 과다(11000개 가량 데이터 중 결측치 10000개 이상), 더 적합한 장소 관련 칼럼 여럿 보유.
    - BRNO : 점포/사업자 식별에 사용되기에 주제 상 불필요.
    - ROAD_NM_ADDR, LOTNO_ADDR : 방문지 데이터가 따로 존재하기에 중복을 피하기 위해 제거
    - ROAD_NM_CD, LOTNO_CD : 위치 식별키. 크기·거리 개념 없음. 외부 행정코드/좌표 매칭용일 때만 의미.
    - PAYMENT_DT : 결제 금액은 중요하지만, 결제 시간까지는 불필요하여 제거
    - CONSUME_HIS_SEQ, CONSUME_HIS_SNO : 소비내역 순번, 부번 코드로, 중복을 피하기 위해 지정된 것이므로 실질적인 영향이 없는 데이터라 제거

- 결측치 처리
    - 'STORE_NM' : 11000개 가량 데이터 중 49개 결측치 존재해 "기타"로 대체
    - ★'PAYMENT_ETC' : 3000개 가량 결측치 존재, "기타"로 대체하나, 해당 데이터 사용시 "기타" 데이터 처리 요망.
